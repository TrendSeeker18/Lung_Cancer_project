**HANDLING MISSING DATA**

```python
# Imputing with the most frequent value
df['Cancer_Stage'] = df['Cancer_Stage'].fillna(df['Cancer_Stage'].mode()[0])
df['Treatment_Type'] = df['Treatment_Type'].fillna(df['Treatment_Type'].mode()[0])

```

```python
print(df.isna().sum())
```

**ANALYSIS**

What is the distribution of numerical columns like "Age" and "Survival_Years"?

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.kdeplot(df['Age'], fill=True, bw_adjust=0.5)  # Adjust bw_adjust for smoothness
plt.title('Age Distribution')
plt.show()

# Distribution for Survival_Years
sns.histplot(df['Survival_Years'], kde=True)
plt.title('Survival Years Distribution')
plt.show()
```

What is the relationship between "Cancer_Stage" and "Survival_Years"

```python
sns.boxplot(x='Cancer_Stage', y='Survival_Years', data=df)
plt.title('Cancer Stage vs Survival Years')
plt.show()
```

Are there any highly correlated features in the dataset?

```python
# Select only numeric columns
numeric_df = df.select_dtypes(include=['number'])

# Generate the heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(numeric_df.corr(), annot=True, fmt='.2f', cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

Are there any outliers in numerical features like "Age" and "Survival_Years"

```python
sns.boxplot(df['Age'])
plt.title('Boxplot for Age')
plt.show()

sns.boxplot(df['Survival_Years'])
plt.title('Boxplot for Survival Years')
plt.show()
```

Does smoking behavior affect lung cancer diagnosis?

```python
sns.countplot(x='Smoker', hue='Lung_Cancer_Diagnosis', data=df)
plt.title('Smoker vs Lung Cancer Diagnosis')
plt.show()
```

What insights can be drawn about healthcare access and treatment types?

```python
sns.countplot(x='Healthcare_Access', hue='Treatment_Type', data=df)
plt.title('Healthcare Access vs Treatment Type')
plt.show()
```


**Summary and Insights Derived**



**Age Distribution:**
Uniform across 20–90 years, indicating no age
bias in the dataset.

**Survival Years Distribution::**
Heavily skewed toward 0, with most patients surviving less than a year, highlighting the need for early
detection.

**Cancer Stage vs. Survival Years:**
The box plot shows that survival years are relatively similar across different cancer stages.
There doesn’t seem to be a strong trend indicating that later-stage cancer significantly reduces survival years based on this visualization.

**Feature Correlation (Heatmap):**
**Strong correlations exist between:**
"Cigarettes per Day" and "Years of Smoking" (0.71) :indicating that
the more years someone smokes, the more cigarettes they tend to consume daily.
"Annual Lung Cancer Deaths" and "Population Size" (0.73): suggesting that larger populations naturally report more deaths.
"Mortality Rate" and "Lung Cancer Prevalence Rate" (0.88):
highlighting that as lung cancer prevalence increases, so does the mortality rate.

**Outliers in Age and Survival Years:**
The boxplot for "Age" does not indicate significant outliers.
The "Survival Years" boxplot shows multiple outliers, meaning some patients have exceptionally long or short survival times.

**Smoking Behavior and Lung Cancer Diagnosis:**
The count plot shows that a significant number of lung cancer diagnoses occur in non-smokers, but smokers still form a considerable portion of diagnosed cases.
This could indicate the presence of other risk factors beyond smoking.

**Healthcare Access vs. Treatment Type:**
Patients with "Poor" healthcare access overwhelmingly receive certain treatments more than those with "Good" access.
This could suggest that poorer healthcare access limits treatment variety or availability. Moreover surgery is preferred over chemotherapy and radiotherapy
